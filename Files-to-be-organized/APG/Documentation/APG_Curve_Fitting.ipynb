```python
import polars as pl
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science', 'notebook', 'grid'])
```

```python
df_L = pl.read_csv('APG-L_Lookup.csv')
df_M = pl.read_csv('APG-M_Lookup.csv')
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_L['Voltage'], df_L['Pressure'], marker='o', linestyle='-', color='b', label='APG-L')
plt.plot(df_M['Voltage'], df_M['Pressure'], marker='o', linestyle='-', color='r', label='APG-M')
plt.yscale('log')
plt.xlabel('Voltage')
plt.ylabel('Pressure')
plt.legend()
plt.title('Voltage vs Pressure')
plt.show()
```
